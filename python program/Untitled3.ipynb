```python
import matplotlib.pyplot as plt
```

```python
x = list(range(100))
y = [t*t for t in x]
plt.scatter(x, y)
plt.title("Square Plot")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.show()
```

```python
x = list(range(-100, 100))
y = [t*t*t for t in x]
plt.scatter(x, y)
plt.title("Cubic Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
